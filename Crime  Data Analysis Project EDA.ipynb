```python
# Extract Year from 'DATE OCC' and analyze crimes by year
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')
data['Year OCC'] = data['DATE OCC'].dt.year
data['Year OCC'].value_counts().sort_index()
```

```python
# Set up the style for the plots
sns.set(style="whitegrid")
```

```python
# Plot Crimes by Year
plt.figure(figsize=(8, 5))
data['Year OCC'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Crimes by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
# Analyze crimes by time of day (categorize into time buckets)
data['TIME OCC'] = pd.to_datetime(data['TIME OCC'], format='%H:%M', errors='coerce').dt.hour
def categorize_time(hour):
    if pd.isnull(hour):
        return 'Unknown'
    elif 0 <= hour < 6:
        return 'Night (12 AM - 6 AM)'
    elif 6 <= hour < 12:
        return 'Morning (6 AM - 12 PM)'
    elif 12 <= hour < 18:
        return 'Afternoon (12 PM - 6 PM)'
    else:
        return 'Evening (6 PM - 12 AM)'

data['Time Category'] = data['TIME OCC'].apply(categorize_time)
data['Time Category'].value_counts()
```

```python
# Plot Crimes by Time of Day Category
plt.figure(figsize=(8, 5))
data['Time Category'].value_counts().plot(kind='bar', color='salmon')
plt.title('Crimes by Time of Day', fontsize=16)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
